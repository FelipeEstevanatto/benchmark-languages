```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Carregar os datasets
C = pd.read_csv("./C.csv", sep=",")
CPP = pd.read_csv("./C++.csv", sep=",")
Java = pd.read_csv("./Java.csv", sep=",")

# Função para plotar gráfico de barras para uma métrica específica (média, moda ou mediana)
def plot_metric(data, metric, metric_name):
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Test', y='Execution Time (ms)', hue='Dataset Size', data=data, errorbar=None, estimator=metric, edgecolor="black")
    plt.title(f'{metric_name} de Tempo de Execução por Algoritmo (Linguagem: {data["Language"].iloc[0]})')
    plt.xlabel('Algoritmo')
    plt.ylabel(f'Tempo de Execução {metric_name} (ms)')
    plt.xticks(rotation=45, ha='right')
    plt.yscale('log')  # Define a escala logarítmica para o eixo y
    plt.legend(title='Tamanho do Dataset')
    plt.tight_layout()
    plt.show()

# Define the metrics
metrics = {
    'Média': np.mean,
    #'Mediana': np.median,
    #'Moda': lambda x: x.mode()[0]
}

# Define the order of the categories
dataset_sizes = ['1k_parc', '10k_parc', '100k_parc', '500k_parc', '1000k_parc', '1k_decreasing', '10k_decreasing', '100k_decreasing', '500k_decreasing', '1000k_decreasing', '1k_random', '10k_random', '100k_random', '500k_random', '1000k_random']

# Calculate each metric for each language
for df, language in zip([C, CPP, Java], ['C', 'C++', 'Java']):
    df['Language'] = language

        # Convert 'Dataset Size' to a categorical data type
    df['Dataset Size'] = pd.Categorical(df['Dataset Size'], categories=dataset_sizes, ordered=True)

    # Sort the DataFrame by 'Dataset Size'
    df.sort_values(['Test', 'Dataset Size'], inplace=True)
    # Filter the DataFrame
    for metric_name, metric_func in metrics.items():
        plot_metric(df, metric_func, metric_name)
```
